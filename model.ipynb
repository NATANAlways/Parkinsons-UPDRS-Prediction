**Dataset Description:**

  This is Parkinsonâ€™s Telemonitoring Data
  ,the dataset contains 5,875 samples and 22 features, representing clinical and voice measurements of patients with Parkinsonâ€™s disease.

a. The Features are:

    age, sex, test_time, Jitter(%), Jitter(Abs), Jitter:RAP, Jitter:PPQ5, Jitter:DDP, Shimmer, Shimmer(dB), Shimmer:APQ3, Shimmer:APQ5, Shimmer:APQ11, Shimmer:DDA, NHR, HNR, RPDE, DFA, PPE, motor_UPDRS, total_UPDRS

The dataset includes the following independent variables (features):  

- **subject#** â€“ Unique identifier for each patient.  
- **age** â€“ Age of the patient.  
- **sex** â€“ Gender (0 = Male, 1 = Female).  
- **test_time** â€“ Elapsed time since the baseline test.  
- **Jitter(%)** â€“ Measure of frequency variation in voice.  
- **Jitter(Abs)** â€“ Absolute measure of frequency perturbation.  
- **Jitter:RAP** â€“ Relative average perturbation.  
- **Jitter:PPQ5** â€“ Five-point period perturbation quotient.  
- **Jitter:DDP** â€“ Average absolute difference of differences between consecutive periods.  
- **Shimmer** â€“ Amplitude variation in voice signal.  
- **Shimmer(dB)** â€“ Shimmer value measured in decibels.  
- **Shimmer:APQ3** â€“ Three-point amplitude perturbation quotient.  
- **Shimmer:APQ5** â€“ Five-point amplitude perturbation quotient.  
- **Shimmer:APQ11** â€“ Eleven-point amplitude perturbation quotient.  
- **Shimmer:DDA** â€“ Average absolute difference of differences between consecutive amplitudes.  
- **NHR** â€“ Noise-to-Harmonics Ratio (higher values indicate more noise in the voice).  
- **HNR** â€“ Harmonics-to-Noise Ratio (higher values indicate a clearer voice signal).  
- **RPDE** â€“ Recurrence period density entropy (used to measure signal complexity).  
- **DFA** â€“ Detrended fluctuation analysis (used for long-range correlation measurement).  
- **PPE** â€“ Pitch period entropy (measures voice signal stability).  


**b.Target Values:**


The dataset has two target values (dependent variables):  

- **motor_UPDRS** â€“ Unified Parkinsonâ€™s Disease Rating Scale (UPDRS) score for motor function.  
- **total_UPDRS** â€“ Overall UPDRS score, including motor and non-motor symptoms.

**Separating target values and features**

```python
# the y value (dependent values)
y = df[['total_UPDRS']]
y.shape
```

```python
# the x values: independent values
x = df.drop(['total_UPDRS', 'motor_UPDRS', 'subject#'], axis=1)
x.shape
```

```python
# information of y values
y.info()
```

```python
# information of x values
x.info()
```

```python
# splitting the dataset as train and test values
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
# Displaying the test and train dataset
print("x_train: ", x_train.shape)
print("x_test: ", x_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

```python
x_train.head()
```

**Linear Regression**

```python
# import linear regression model
from sklearn.linear_model import LinearRegression
# import evaluating the model
from sklearn.metrics import mean_squared_error, r2_score

# Create a linear regression model
model = LinearRegression()
```

```python
# train the model
model.fit(x_train, y_train)
```

```python
# make prediction using the test data
y_pred_LR = model.predict(x_test)
```

```python
y_pred_LR
```

```python
# the linear Regression model coeffieceint and intercept
print(f"Coefficient: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
# model evaluation
# Mean Square Error
mse_LR = mean_squared_error(y_test, y_pred_LR)
# R - squared value
r2_LR = r2_score(y_test, y_pred_LR)
```

```python
# Displaying the evaluation results
print(f"Mean Squared Error: {mse_LR}")
print(f"Root Mean Squared Error: {np.sqrt(mse_LR)}")
print(f"R-squared: {r2_LR}")
```

**Random Forest Regression**

```python
# for model building the Random Forest model
from sklearn.ensemble import RandomForestRegressor
```

```python
# Initializing the Random Forest Regression model
model = RandomForestRegressor(n_estimators=100,random_state=42)
# train
model.fit(x_train,y_train)
# predict
y_pred_RF = model.predict(x_test)
```

```python
# model evaluation
# Mean Square Error
mse_RF = mean_squared_error(y_test, y_pred_RF)
# R - squared value
r2_RF = r2_score(y_test, y_pred_RF)
```

```python
# Displaying the evaluation results
print(f"Mean Squared Error: {mse_RF}")
print(f"Root Mean Squared Error: {np.sqrt(mse_RF)}")
print(f"R-squared: {r2_RF}")
```

**Support Vector Regression**

```python
# import support vector regression model
from sklearn.svm import SVR
# for scaling
from sklearn.preprocessing import StandardScaler
```

```python
# SVR model
model = SVR(kernel = "rbf")
# train the model
model.fit(x_train, y_train)
# make prediction on the test data
y_pred_SVR = model.predict(x_test)
```

```python
# model evaluation
# Mean Square Error
mse_SVR = mean_squared_error(y_test, y_pred_SVR)
# R - squared value
r2_SVR = r2_score(y_test, y_pred_SVR)
```

```python
# Displaying the evaluation results
print(f"Mean Squared Error: {mse_SVR}")
print(f"Root Mean Squared Error: {np.sqrt(mse_SVR)}")
print(f"R-squared: {r2_SVR}")
```
